```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
import torch
from transformers import TrainingArguments, Trainer
from transformers import EarlyStoppingCallback
from transformers import AutoTokenizer, FlaubertForSequenceClassification
from sklearn import metrics
```

```python
# Function to open data
def open_data(name):
    return pd.read_csv(name, index_col=0)
```

```python
# Read data
train_data = open_data("Path to your test_set")
```

```python
# Merge sentences and rationals
train_data["Features"] = train_data["Phrases"] + " " + train_data["Rationals"]
```

```python
# Define pretrained tokenizer and model

model_name = 'flaubert/flaubert_base_cased'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = FlaubertForSequenceClassification.from_pretrained(model_name, num_labels=11)
```

```python
# Preprocess data
X = list(train_data["Features"])
y = list(train_data["Labels"])
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
X_train_tokenized = tokenizer(X_train, padding=True, truncation=True, max_length=512)
X_val_tokenized = tokenizer(X_val, padding=True, truncation=True, max_length=512)
```

```python
# Create torch dataset
class Dataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels=None):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        if self.labels:
            item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.encodings["input_ids"])
```

```python
train_dataset = Dataset(X_train_tokenized, y_train)
val_dataset = Dataset(X_val_tokenized, y_val)
```

```python
type(train_dataset)
```

```python
# ----- 2. Fine-tune pretrained model -----#
# Define Trainer parameters
def compute_metrics(pred, labels):
    #pred, labels = p
    pred = np.argmax(pred, axis=1)

    accuracy = accuracy_score(y_true=labels, y_pred=pred)
    recall = recall_score(y_true=labels, y_pred=pred,  average='weighted')
    precision = precision_score(y_true=labels, y_pred=pred, average='weighted')
    f1 = f1_score(y_true=labels, y_pred=pred, average='weighted')

    return {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1}
```

```python
# Define Trainer
args = TrainingArguments(
    output_dir="output",
    evaluation_strategy="steps",
    eval_steps=500,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=100,
    save_steps=1000,
    seed=0,
    load_best_model_at_end=True,
)
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=5)],
)
```

```python
# Train pre-trained model
trainer.train()
```

```python
# Upload your test data
test_data = open_data("Path to your test_set")
```

```python
test_data
```

```python
# ----- 3. Predict -----#
# Load test data
X_test = list(test_data["text-fr"])
y_test = list(test_data["labels"])
X_test_tokenized = tokenizer(X_test, padding=True, truncation=True, max_length=512)
```

```python
# Create torch dataset
test_dataset = Dataset(X_test_tokenized)
```

```python
# Load trained mode
model_path = "output/checkpoint-1000"
model = FlaubertForSequenceClassification.from_pretrained(model_path, num_labels=11)
```

```python
# Define test trainer
test_trainer = Trainer(model)
```

```python
# Make prediction
raw_pred, _, _ = test_trainer.predict(test_dataset)
```

```python
#Get metrics
compute_metrics(raw_pred, y_test)
```
